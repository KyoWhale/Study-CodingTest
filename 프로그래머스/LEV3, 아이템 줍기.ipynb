```python
from collections import deque

dx, dy = [0,1,0,-1], [1,0,-1,0]

def solution(rectangles, charX, charY, itemX, itemY):
    answer = 0
    MAX = 102

    field = [[5] * MAX for _ in range(MAX)]
    for rectangle in rectangles:
        x1, y1, x2, y2 = map(lambda x: x* 2, rectangle)
        for i in range(y1, y2+1):
            for j in range(x1, x2+1):
                if y1 < i < y2 and x1 < j < x2:
                    field[i][j] = 0
                elif field[i][j] != 0:
                    field[i][j] = 1

    q = deque()
    q.append([charY * 2, charX * 2])
    visited = [[0] * MAX for _ in range(MAX)]
    visited[charY * 2][charX * 2] = 1
    while q:
        y, x = q.popleft()
        if x == itemX * 2 and y == itemY * 2:
            answer = (visited[y][x] - 1) // 2
            break

        for i in range(4):
            ny, nx = y + dy[i], x + dx[i]
            if visited[ny][nx] == 0 and field[ny][nx] == 1:
                q.append([ny,nx])
                visited[ny][nx] = visited[y][x] + 1

    return answer
solution([[1,1,7,4],[3,2,5,5],[4,3,6,9],[2,6,8,8]],1,3,7,8)
solution([[1,1,8,4],[2,2,4,9],[3,6,9,8],[6,3,7,7]],9,7,6,1)
```
